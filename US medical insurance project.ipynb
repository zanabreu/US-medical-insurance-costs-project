```python
def stats_variable(variable):
    mean_variable=insurance_df[variable].mean()
    print("The average " + str(variable) + " is:")
    print(mean_variable)
    mode_variable=insurance_df[variable].mode()
    print("The more frequent " + str(variable) + " is:")
    print(mode_variable)
    max_variable=insurance_df[variable].max()
    print("The maximum " + str(variable) + " is:")
    print(max_variable)
    min_variable=insurance_df[variable].min()
    print("The minimum " + str(variable) + " is:")
    print(min_variable)
```

```python
# CORRELATION WEIGHT AND SMOKING

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Assuming insurance_df is your dataframe with 'smoker' and 'bmi' columns

# 1. Boxplot to compare BMI distributions between smokers and non-smokers
plt.figure(figsize=(5, 2.5))
sns.boxplot(x='smoker', y='bmi', data=insurance_df)
plt.title('BMI of smokers and non-smokers')
plt.xlabel("Smoking status")
plt.xticks([0, 1], ["Smoker", "Non-smoker"])
plt.ylabel('BMI')
plt.show()

# 2. Calculate summary statistics for each group
smoker_stats = insurance_df.groupby('smoker')['bmi'].describe()
print("BMI statistics by smoking status:")
print(smoker_stats)

# 3. Perform t-test to check if difference is statistically significant
smokers_bmi = insurance_df[insurance_df['smoker'] == 'yes']['bmi']
non_smokers_bmi = insurance_df[insurance_df['smoker'] == 'no']['bmi']
t_stat, p_value = stats.ttest_ind(smokers_bmi, non_smokers_bmi)
print(f"\nT-test results: t-statistic = {t_stat:.4f}, p-value = {p_value:.4f}")
print(f"Statistically significant difference: {p_value < 0.05}")

# 4. Violin plot (shows distribution shape)
plt.figure(figsize=(5, 2.5))
sns.violinplot(x='smoker', y='bmi', data=insurance_df)
plt.title('BMI of smokers and non-smokers')
plt.xlabel("Smoking status")
plt.xticks([0, 1], ["Smoker", "Non-smoker"])
plt.show()

# 5. Calculate correlation between smoking status (converted to numeric) and BMI
# Convert 'yes'/'no' to 1/0 for correlation calculation
insurance_df['smoker_numeric'] = insurance_df['smoker'].map({'yes': 1, 'no': 0})
correlation = insurance_df['smoker_numeric'].corr(insurance_df['bmi'])
print(f"\nCorrelation coefficient between smoking status and BMI: {correlation:.4f}")
```

```python
#CORRELATION AGE AND SMOKING

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Assuming insurance_df is your dataframe with 'smoker' and 'bmi' columns

# 1. Boxplot to compare BMI distributions between smokers and non-smokers
plt.figure(figsize=(5, 2.5))
sns.boxplot(x='smoker', y='age', data=insurance_df)
plt.title('Age of smokers and non-smokers')
plt.xlabel("Smoking status")
plt.xticks([0, 1], ["Smoker", "Non-smoker"])
plt.ylabel('Age')
plt.show()

# 2. Calculate summary statistics for each group
smoker_stats_per_age = insurance_df.groupby('smoker')['age'].describe()
print("Age statistics by smoking status:")
print(smoker_stats_per_age)

# 3. Perform t-test to check if difference is statistically significant
smokers_age= insurance_df[insurance_df['smoker'] == 'yes']['age']
non_smokers_age = insurance_df[insurance_df['smoker'] == 'no']['age']
t_stat, p_value = stats.ttest_ind(smokers_age, non_smokers_age)
print(f"\nT-test results: t-statistic = {t_stat:.4f}, p-value = {p_value:.4f}")
print(f"Statistically significant difference: {p_value < 0.05}")

# 4. Violin plot (shows distribution shape)
plt.figure(figsize=(5, 2.5))
sns.violinplot(x='smoker', y='age', data=insurance_df)
plt.title('Age of smokers and non-smokers')
plt.xlabel("Smoking status")
plt.xticks([0, 1], ["Smoker", "Non-smoker"])
plt.show()

# 5. Calculate correlation between smoking status (converted to numeric) and BMI
# Convert 'yes'/'no' to 1/0 for correlation calculation
insurance_df['smoker_numeric'] = insurance_df['smoker'].map({'yes': 1, 'no': 0})
correlation = insurance_df['smoker_numeric'].corr(insurance_df['age'])
print(f"\nCorrelation coefficient between smoking status and age: {correlation:.4f}")
```

```python
#CORRELATION Charges AND SMOKING

import scipy.stats as stats

# Boxplot to compare charges between smokers and non-smokers
plt.figure(figsize=(5, 2.5))
sns.boxplot(x='smoker', y='charges', data=insurance_df)
plt.title('Charges of smokers vs non-smokers')
plt.xticks([0, 1], ["Smoker", "Non-smoker"])
plt.xlabel("Smoking Status")
plt.ylabel('Charges ($)')
plt.show()


plt.figure(figsize=(5, 2.5))
# Add individual data points for better visualization
sns.stripplot(x="smoker", y="charges", data=insurance_df, size=3, color=".3", linewidth=0, alpha=0.5)
plt.xticks([0, 1], ["Smoker", "Non-smoker"])
plt.xlabel("Smoking Status")
plt.ylabel("Charges ($)")
plt.title("Charges of smokers vs non-smokers")
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
plt.clf()


#Violin plot (shows distribution shape)
plt.figure(figsize=(5, 2.5))
sns.violinplot(x='smoker', y='charges', data=insurance_df)
plt.xticks([0, 1], ["Smoker", "Non-smoker"])
plt.xlabel("Smoking Status")
plt.ylabel("Charges ($)")
plt.title('Charges in smokers vs non-smokers')
plt.show()

# Calculate mean charges for each group
non_smokers_mean = insurance_df[insurance_df["smoker"] == "no"]["charges"].mean()
smokers_mean = insurance_df[insurance_df["smoker"] == "yes"]["charges"].mean()
print("\nNon-smokers mean charges: $" + str(round(non_smokers_mean,2)))
print("\nSmokers mean charges: $" + str(round(smokers_mean,2)))

#Calculate difference of means of charges paid by smokers and non-smokers

smokers_mean_charges = insurance_df[insurance_df["smoker"]=="yes"]["charges"].mean()
nonsmokers_mean_charges = insurance_df[insurance_df["smoker"]=="no"]["charges"].mean()
how_much_more_smokers_pay = smokers_mean_charges - nonsmokers_mean_charges
fold_higher_charge_smokers = smokers_mean_charges / nonsmokers_mean_charges
print("\nSmokers pay in average " + str(round(how_much_more_smokers_pay,2)) + " dollars more than non-smokers.")
print("\nSmokers pay " + str(round(fold_higher_charge_smokers,2)) + " times more insurance than non-smokers.")

#Min and max

smokers_max_charges = insurance_df[insurance_df["smoker"]=="yes"]["charges"].max()
smokers_min_charges = insurance_df[insurance_df["smoker"]=="yes"]["charges"].min()
nonsmokers_max_charges = insurance_df[insurance_df["smoker"]=="no"]["charges"].max()
nonsmokers_min_charges = insurance_df[insurance_df["smoker"]=="no"]["charges"].min()
print("\nSmokers pay a minimum of " + str(round(smokers_min_charges,2)) + " and a maximum of " + str(round(smokers_max_charges,2)) + ", while non-smokers pay a minimum of " + str(nonsmokers_min_charges) + " and a maximum of " + str(nonsmokers_max_charges) + ".")

#Variation of charges in smokers and non-smokers

var_charges_smokers = smokers_max_charges - smokers_min_charges
var_charges_nonsmokers = nonsmokers_max_charges - nonsmokers_min_charges
fold_var_smokers_nonsmokers = var_charges_smokers /var_charges_nonsmokers 
print("\nSmokers charges fluctuate " + str(round(fold_var_smokers_nonsmokers,1)) + " times more than charges of non-smokers.")

# Perform t-test to check if difference is statistically significant
smokers_charges= insurance_df[insurance_df['smoker'] == 'yes']['charges']
non_smokers_charges = insurance_df[insurance_df['smoker'] == 'no']['charges']
t_stat, p_value = stats.ttest_ind(smokers_charges, non_smokers_charges)
print(f"\nT-test results: t-statistic = {t_stat:.4f}, p-value = {p_value:.4f}")
print(f"\nStatistically significant difference between smokers and non-smokers charges: {p_value < 0.05}")
```
